```python
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
from diffusers.utils import export_to_video

pipe = DiffusionPipeline.from_pretrained("ali-vilab/text-to-video-ms-1.7b", torch_dtype = torch.float16, variant="fp16")
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe.enable_model_cpu_offload()
```

```python
prompt = "a lion running to prey"
video_frames = pipe(prompt, num_inference_steps = 25).frames
# Reshape or select the correct frames from the array
video_frames = video_frames[0]  # Select the first (and only) batch

video_path = export_to_video(video_frames)
video_name = video_path.replace('/tmp','')
print('Name', video_name)
torch.cuda.empty_cache()
```

```python
# prompt: to make this work as user input prompt

import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
from diffusers.utils import export_to_video
!pip install diffusers transformers accelerate

pipe = DiffusionPipeline.from_pretrained("ali-vilab/text-to-video-ms-1.7b", torch_dtype = torch.float16, variant="fp16")
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe.enable_model_cpu_offload()

prompt = input("Enter a prompt for video generation: ")

video_frames = pipe(prompt, num_inference_steps = 25).frames
# Reshape or select the correct frames from the array
video_frames = video_frames[0]  # Select the first (and only) batch

video_path = export_to_video(video_frames)
video_name = video_path.replace('/tmp','')
print('Name', video_name)
torch.cuda.empty_cache()
```
